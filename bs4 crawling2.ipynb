```python
import pandas as pd
import time
import requests
from bs4 import BeautifulSoup as bs
import ast
import re
from kss import split_sentences
```

```python
def get_대출사기형_info():
    
    
        
    list_url = []
    list_content = []

    for page in range(19):
            
            main_url = f'https://www.fss.or.kr/fss/bbs/B0000206/list.do?menuNo=200690&bbsId=&cl1Cd=&pageIndex={page+1}&sdate=&edate=&searchCnd=1&searchWrd='
        
            response = requests.get(main_url)
            soup = bs(response.text, 'html.parser')
        
            current = soup.select(f'#content > div.bd-list > table > tbody > tr > td.title > a')
        
            
            for url in current:
                list_url.append(url.attrs['href'])
    
    print(list_url)
    
    count = 0 
    for url in list_url:
        print(count)
        url = list_url[count]
        print(url)
        response = requests.get("https://www.fss.or.kr"+url)
        soup = bs(response.text, 'html.parser')
        count = count + 1
        
        cur_content = []    
        current = soup.select('#content > div.bd-view > div > div > p')

        
        if len(current) == 0:
            list_content.append(cur_content)
            continue
        
        for sentence in current:
            cur_content.append(sentence.text)
        print(cur_content)
        list_content.append(cur_content)
        print('게시글내용append')
  
    print(list_content)
    
    
    voicephishing_대출사기형 = pd.DataFrame({'url':list_url, 'content':list_content})
    
    return voicephishing_대출사기형
    
    
    
  
    
```

```python
import pandas as pd
import time
import requests
from bs4 import BeautifulSoup as bs

def get_수사기관사칭형_info():
    
    
        
    list_url = []
    list_content = []

    for page in range(23):
            
            main_url = f'https://www.fss.or.kr/fss/bbs/B0000207/list.do?menuNo=200691&bbsId=&cl1Cd=&pageIndex={page+1}&sdate=&edate=&searchCnd=1&searchWrd='
        
            response = requests.get(main_url)
            soup = bs(response.text, 'html.parser')
        
            current = soup.select(f'#content > div.bd-list > table > tbody > tr > td.title > a')
            
            for url in current:
                list_url.append(url.attrs['href'])
    
    print(list_url)
    
    count = 0 
    for url in list_url:
        print(count)
        url = list_url[count]
        print(url)
        response = requests.get("https://www.fss.or.kr"+url)
        soup = bs(response.text, 'html.parser')
        count = count + 1
        
        cur_content = []    
        current = soup.select('#content > div.bd-view > div > div > p')
        
        if len(current) == 0:
            list_content.append(cur_content)
            continue
        
        for sentence in current:
            cur_content.append(sentence.text)
        print(cur_content)
        list_content.append(cur_content)
        print('게시글내용append')
  
    print(list_content)
    
    
    voicephishing_수사기관사칭형 = pd.DataFrame({'url':list_url, 'content':list_content})
    
    return voicephishing_수사기관사칭형
    
    
    
  
    
```

```python
voicephishing_수사기관사칭형 = get_수사기관사칭형_info()
```

```python
voicephishing_수사기관사칭형.head(10)
```

```python
voicephishing_대출사기형_text = list(voicephishing_대출사기형['content'])
voicephishing_대출사기형_text = [i for i in voicephishing_대출사기형_text if i != '[]']
len_대출사기형_text = len(voicephishing_대출사기형_text)
print(f' 대출사기형 보이스피싱 텍스트 건수: {len_대출사기형_text}')

voicephishing_수사기관사칭형_text = list(voicephishing_수사기관사칭형['content'])
voicephishing_수사기관사칭형_text = [i for i in voicephishing_수사기관사칭형_text if i != '[]']
len_수사기관사칭형_text = len(voicephishing_수사기관사칭형_text)
print(f' 수사기관사칭형 보이스피싱 텍스트 건수: {len_수사기관사칭형_text}')


# 문자열의 리스트화
for i in range(len_대출사기형_text):
    cur_text = voicephishing_대출사기형_text[i]
    cur_text = ast.literal_eval(cur_text)
    voicephishing_대출사기형_text[i] = cur_text
    
for i in range(len_수사기관사칭형_text):
    cur_text = voicephishing_수사기관사칭형_text[i]
    cur_text = ast.literal_eval(cur_text)
    voicephishing_수사기관사칭형_text[i] = cur_text
```

```python
# 피해자/사기범 명칭 제거
for i in range(len_대출사기형_text):
    cur_text = voicephishing_대출사기형_text[i]
    
    for j in range(len(cur_text)):
        # 피해자 명칭 제거
        if ('피해자:' in cur_text[j]) or ('피해자 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'피해자:', '', cur_text[j])
            cur_text[j] = re.sub(r'피해자 : ', '', cur_text[j])
            voicephishing_대출사기형_text[i][j] = cur_text[j]
            continue
            
        # 사기범 명칭 제거
        if ('사기범:' in cur_text[j]) or ('사기범 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'사기범:', '', cur_text[j])
            cur_text[j] = re.sub(r'사기범 : ', '', cur_text[j])
            voicephishing_대출사기형_text[i][j] = cur_text[j]
                
                
for i in range(len_수사기관사칭형_text):
    cur_text = voicephishing_수사기관사칭형_text[i]
    
    for j in range(len(cur_text)):
        # 피해자 명칭 제거
        if ('피해자:' in cur_text[j]) or ('피해자 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'피해자:', '', cur_text[j])
            cur_text[j] = re.sub(r'피해자 : ', '', cur_text[j])
            voicephishing_수사기관사칭형_text[i][j] = cur_text[j]
            continue
            
        # 사기범 명칭 제거
        if ('사기범:' in cur_text[j]) or ('사기범 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'사기범:', '', cur_text[j])
            cur_text[j] = re.sub(r'사기범 : ', '', cur_text[j])
            voicephishing_수사기관사칭형_text[i][j] = cur_text[j]
```

```python
# 문장 분리
for i in range(len_대출사기형_text):
    cur_text = voicephishing_대출사기형_text[i]
    splited_text = []
    
    for j in range(len(cur_text)):       
        # 문장 분리 후 길이 저장
        splited = split_sentences(cur_text[j])
        len_splited = len(splited)
        
        # 길이가 1이면 바로 추가함
        if len_splited == 1:
            splited_text.append(cur_text[j])
        # 길이가 1이 아니면 나눠서 추가함
        else:
            for k in range(len_splited):
                splited_text.append(splited[k])
            
    # 분리한 문장으로 대체
    voicephishing_대출사기형_text[i] = splited_text
    
    
for i in range(len_수사기관사칭형_text):
    cur_text = voicephishing_수사기관사칭형_text[i]
    splited_text = []
    
    for j in range(len(cur_text)):       
        # 문장 분리 후 길이 저장
        splited = split_sentences(cur_text[j])
        len_splited = len(splited)
        
        # 길이가 1이면 바로 추가함
        if len_splited == 1:
            splited_text.append(cur_text[j])
        # 길이가 1이 아니면 나눠서 추가함
        else:
            for k in range(len_splited):
                splited_text.append(splited[k])
            
    # 분리한 문장으로 대체
    voicephishing_수사기관사칭형_text[i] = splited_text
```

```python
# 욕설 제거
for i in range(len_대출사기형_text):
    cur_text = voicephishing_대출사기형_text[i]
    
    for j in range(len(cur_text)):                      
        # 욕설 이후 제거
        if 'xx' in cur_text[j]:
            del voicephishing_대출사기형_text[i][j:]
            break
                   
                
for i in range(len_수사기관사칭형_text):
    cur_text = voicephishing_수사기관사칭형_text[i]
    
    for j in range(len(cur_text)):                     
        # 욕설 이후 제거
        if 'xx' in cur_text[j]:
            del voicephishing_수사기관사칭형_text[i][j:]
            break
```

```python
# 문장당 카운트 횟수
countsent_대출사기형 = 0
for i in range(len_대출사기형_text):
    countsent_대출사기형 += len(voicephishing_대출사기형_text[i])
print(f' 대출사기형 사기범 문장 개수: {countsent_대출사기형}')
    
countsent_수사기관사칭형 = 0
for i in range(len_수사기관사칭형_text):
    countsent_수사기관사칭형 += len(voicephishing_수사기관사칭형_text[i])
print(f' 수사기관사칭형 사기범 문장 개수: {countsent_수사기관사칭형}')
```

```python
# 파일 통합
voicephishing_전체_text = voicephishing_대출사기형_text + voicephishing_수사기관사칭형_text
len_전체_text = len(voicephishing_전체_text)
print(f' 보이스피싱 전체 사건 개수: {len_전체_text}')

countsent_전체 = 0
for i in range(len_전체_text):
    countsent_전체 += len(voicephishing_전체_text[i])
print(f' 전체 문장 개수: {countsent_전체}')
```
